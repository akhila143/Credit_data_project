<span style="font-size: 22px;">**CASE STUDY: GOLD ATLANTIS: REPAYING THE CREDIT**</span>

<p><span style="font-size: 20px; text-decoration: underline;">1. SUMMARY</span></p>
<p><span style="font-size: 15px;">Gold Atlantis, a credit-providing firm, observed a rise in the number of defaulters.Now, the firm is only interested in those consumers who have a lower probability of becoming a defaulter.</span></p>
<p><span style="font-size: 15px;">When Gold Atlantis recieves a credit application, it approves or disapproves the application based on the applicant's profile.</span></p>
<p><span style="font-size: 15px;"> However, there are two types of risks that are associated with the firm decisions:</span></p>
<ul>
  <li><span style="font-size: 15px;">If the applicant is likely to repay the loan, then disapproving the loan may result in a loss of business to the firm.</span></li>
  <li><span style="font-size: 15px;">If the applicant is not likely to repay the loan, that is, if the applicant is a defaulter, then approving the loan may lead to a financial loss to the firm.</span></li>
</ul>

<p><span style="font-size: 15px;">The main aim of this case study is to help firm identify the customers who have a lower probability of becoming a defaulter.</span></p>

<p><span style="font-size: 15px">4.4.1 Checking for null values</p>
<p><span style="font-size: 15px"> Now we need to check if there are any null values in the data and replace them with mode or mean of the data to avoid inconsistencies while analyzing the data.</p>

````python
# Check if there are any null values
data.isnull().sum()
````

<p><span style="font-size: 15px">So there are null values in the data. Replacing the null values with mode if they are categorical and with mean if they are numerical columns.</p>

````python
cat = []
num = []
def cat_or_num(data):
    for i in data.columns:
        if data[i].nunique() <= 30:
            cat.append(i)
        else:
            num.append(i)

cat_or_num(data)

# Filling the null values in categorical columns with mode and the numerical columns with mean. 
for i in data.columns:
    if i in cat:
        data[i].fillna(data[i].mode()[0],inplace = True)
    elif i in num:
        data[i].fillna(data[i].mean(), inplace = True)
````

````python
# Validating if there are any null vales  
data.isnull().sum()
````

<p><span style="font-size: 15px">4.4.2 Checking for duplicate values</p>
<p><span style="font-size: 15px"> Now we need to check if there are any duplicate values in the data and remove them to avoid inconsistencies while analyzing the data.</p>

````python
# Checking if there are any duplicate values and removing them.
duplicates = data[data.duplicated(keep=False)]

if not duplicates.empty:
    data.drop_duplicates(inplace = True)
    print("Removed all the duplicate values from the data")
else:
    print("No duplicate rows found")
````

Defining a function that can give us the graphs that are needed for univariate and bivariate analysis.

````python
def analyse(column,i,j):
    fig, axs = plt.subplots(1,3,figsize=(20,5))

    sb.countplot(x = data[column],ax = axs[0])
    axs[0].set_title(f"Count of people by {column}")
    total1 = len(data[column])
    for p in axs[0].patches:
        height = p.get_height()
        percentage = (height / total1) * 100
        axs[0].annotate(f'{percentage:.0f}%', (p.get_x() + p.get_width() / 2., height + 0.2),
                     ha='center', fontsize=12)

    table1 = data[data[column] == i]
    index = table1['TARGET'].value_counts().index
    values = table1['TARGET'].value_counts().values
    axs[1].pie(x = values, labels = index, autopct = '%0.2f%%' )
    axs[1].set_title(i)

    table2 = data[data[column] == j]
    index = table2['TARGET'].value_counts().index
    values = table2['TARGET'].value_counts().values
    axs[2].pie(x = values, labels = index, autopct = '%0.2f%%' )
    axs[2].set_title(j)

    for a in axs:
        a.set_xticklabels(a.get_xticklabels(), rotation=90)
        
    plt.show()
````

````python
analyse('NAME_CONTRACT_TYPE','Cash loans','Revolving loans')
print("People are taking more cash loans than revolving loans.")
print("People who take revolving loans have less defaulters.")
````

````python
n_values = []
target_0_counts = []
target_1_counts = []

for n in range(1, 11):
    k = data[data['AMT_CREDIT'] == n * data['AMT_INCOME_TOTAL']]
    
    target_0_count = k['TARGET'].value_counts().get(0, 0)
    target_1_count = k['TARGET'].value_counts().get(1, 0)
    
    n_values.append(n)
    target_0_counts.append(target_0_count)
    target_1_counts.append(target_1_count)

n_values = np.array(n_values)
target_0_counts = np.array(target_0_counts)
target_1_counts = np.array(target_1_counts)

total_counts = target_0_counts + target_1_counts
target_0_percentages = (target_0_counts / total_counts) * 100
target_1_percentages = (target_1_counts / total_counts) * 100

bar_width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))

bar1 = ax.bar(n_values, target_0_counts, width=bar_width, label='Count of TARGET=0', alpha=0.7)
bar2 = ax.bar(n_values + bar_width, target_1_counts, width=bar_width, label='Count of TARGET=1', alpha=0.7)

ax.set_title('Bar Plot of TARGET Counts vs. n')
ax.set_xlabel('n(Number of times the income is credit)')
ax.set_ylabel('Count')

for n, perc_0, perc_1 in zip(n_values, target_0_percentages, target_1_percentages):
    ax.text(n - bar_width/2, target_0_counts.max() + 10, f'{perc_0:.1f}%', ha='center', va='bottom')
    ax.text(n + bar_width/2, target_1_counts.max() + 10, f'{perc_1:.1f}%', ha='center', va='bottom')

ax.legend()

plt.show()
````

As n (Number of times the income is credit) increases the number of defaulters increases.

#### Univariate and Bivariate Analysis Insights:

##### 1. Loan Type and Default Rates:
- Cash loans have a higher percentage of defaulters compared to revolving loans, indicating that borrowers with revolving loans are more likely to meet their repayment obligations. This could be due to the more flexible nature of revolving credit lines.

##### 2. Gender and Default Rates:
- Males have a higher default rate compared to females, suggesting that women are more reliable borrowers when it comes to loan repayments. This might be attributed to various factors, such as income stability or financial responsibility.

##### 3. Family Size and Default Rates:
- Individuals with no children tend to default less compared to those with a single child. This implies that having no children may provide borrowers with more financial stability, making them better candidates for loans.

##### 4. Employment Status and Default Rates:
- Working individuals have a lower default rate than pensioners and the unemployed. This indicates that a stable source of income positively impacts loan repayment reliability.

##### 5. Educational Background and Default Rates:
- Borrowers with secondary education have a higher default rate compared to those with higher education. This suggests that individuals with higher educational qualifications might have better financial management skills, reducing their likelihood of default.

##### 6. Occupational Skill Level and Default Rates:
- Unskilled workers have a higher default rate compared to skilled workers. This could be due to income disparities and the financial stability associated with higher-skilled occupations.

##### 7. Document Requirements and Default Rates:
- Borrowers with fewer than or equal to one document have a higher default rate than those with two or more documents. This suggests that providing more documentation during the loan application process might be associated with lower default risk, possibly indicating greater transparency in financial matters.

##### 8. Marital Status and Default Rates:
- Single individuals have a higher default rate compared to married individuals and widows. This could be because married individuals may have dual sources of income or financial support systems that reduce their default risk.

##### 9. Car Ownership and Default Rates:
- The similarity in default rates between those who do not own a car and those who do suggests that car ownership may not be a significant factor in determining loan default risk.

##### 10. Home Ownership and Default Rates:

##### 12. Family Size  and Default Rates:

#### Multivariate Analysis Insights:

##### 1. High Credit vs. Income in Cash Loans:
   - People taking cash loans are often extended credit amounts that exceed their income levels, leading to a higher incidence of defaulters in this category. Lenders should reconsider the credit limits they offer to align with borrowers' income capacity.

#####  2. Gender Disparity in Repayment:
   - While men and women are offered similar credit amounts, females tend to have a better repayment record. This indicates a potential need for further analysis to understand if there are specific schemes or support systems in place benefiting female borrowers.

#####  3. Credit and Company Advantages:
   - Borrowers with higher education levels often receive more credit than their income suggests, yet they have lower default rates. This suggests the need to investigate whether these individuals benefit from workplace advantages or financial education programs that enhance their ability to manage debt effectively.

#####  4. Default Risk of Pensioners:
   - Pensioners tend to take out smaller loans compared to other groups, resulting in a lower number of defaulters. This implies that pensioners are cautious borrowers, possibly due to their fixed income. Lenders should consider this group's financial stability when evaluating loan applications.

#####  5. Effect of Document Submission:
   - Borrowers submitting more than two documents, despite having credit amounts significantly exceeding their income, have a lower default rate. This highlights the importance of analyzing the types and significance of the documents provided during the loan application process to identify factors contributing to their lower default rates.

#####  6. Impact of Group Accompaniment:
   - Borrowers who are accompanied by a group of people receive lower credit amounts, which may be attributed to collective decision-making or group accountability. Consequently, this group experiences fewer defaulters, suggesting that social support may influence repayment behavior positively.

#####  7. Widows and Conservative Borrowing:
   - Widows tend to borrow smaller amounts compared to other groups, resulting in fewer defaulters. This suggests that individuals in certain life circumstances, such as widowhood, may approach borrowing conservatively, resulting in better repayment behavior.

#####  8. Income vs. Credit Multiplier Effect:
   - There is an observed increase in the number of defaulters as the number of times income is exceeded by the credit amount (credit multiplier) increases. This highlights the importance of setting responsible lending limits and assessing the risk associated with higher credit multipliers.

In summary, these insights indicate that various demographic and financial factors can influence the likelihood of loan default. Lenders should consider these factors when evaluating loan applications and setting interest rates to mitigate default risk effectively.
